```python
import dbscan_tensorflow as dbtf
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
blobs_data, blobs_labels = make_blobs(n_samples=1000, n_features=2, centers=5, cluster_std=0.3, random_state=101)
```

```python
plt.scatter(blobs_data[:,0], blobs_data[:,1], c=blobs_labels)
plt.show()
```

```python
blobs_results = dbtf.run(blobs_data, epsilon=1, min_points=4)

plt.scatter(blobs_data[:,0], blobs_data[:,1], c=blobs_results.flatten())
plt.show()
```
